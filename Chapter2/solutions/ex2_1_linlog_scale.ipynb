# Comparing both scales

When comparing linear and logarithmic scales on a financial chart, the key difference lies in how price movements are visually represented. A linear scale displays equal vertical spacing for equal absolute price changes, meaning a move from $10 to $20 looks the same as one from $100 to $110, even though their percentage gains differ greatly. In contrast, a logarithmic scale spaces the chart based on percentage changes, so a 100% increaseâ€”from $10 to $20 or $100 to $200â€”appears equally large. This makes log scales more appropriate for analyzing long-term investments or assets with exponential growth, as they offer a clearer view of relative performance across different price levels. While both charts may look similar over short periods or small price ranges, their interpretations diverge significantly when assessing trends over time or across wide price fluctuations.
